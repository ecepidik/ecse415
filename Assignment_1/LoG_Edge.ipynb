```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('woman.png',)

# converting to gray scale
imgLoG = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# remove noise
imgLoG = cv2.GaussianBlur(imgLoG,(5,5),0)

# convolute with proper kernels
imgLoG = cv2.Laplacian(imgLoG,cv2.CV_64F)

plt.figure(figsize=(20,20))
plt.subplot(121), plt.imshow(img, cmap="gray")
plt.title("Image"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(imgLoG, cmap="gray")
plt.title("LoG_edge"), plt.xticks([]), plt.yticks([])
plt.show()
```

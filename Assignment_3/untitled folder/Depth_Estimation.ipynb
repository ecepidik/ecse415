```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

imgL = cv2.imread('cones/im2.png')
imgR = cv2.imread('cones/im6.png')

dispL = cv2.imread('cones/disp2.png')
dispR = cv2.imread('cones/disp6.png')

# resize images for fast processing
imgL = cv2.resize(imgL, None, fx=0.3, fy=0.3)
imgR = cv2.resize(imgR, None, fx=0.3, fy=0.3)

dispL = cv2.resize(dispL, None, fx=0.3, fy=0.3)
dispR = cv2.resize(dispR, None, fx=0.3, fy=0.3)

# convert to grayscale
imgL_gray = cv2.cvtColor(imgL, cv2.COLOR_BGR2GRAY)
imgR_gray = cv2.cvtColor(imgR, cv2.COLOR_BGR2GRAY)     

dispL_gray = cv2.cvtColor(dispL, cv2.COLOR_BGR2GRAY)
dispR_gray = cv2.cvtColor(dispR, cv2.COLOR_BGR2GRAY)    

blockSizes = [5,9,15]
numDisparities = [16,32,48]

i=1
for block in blockSizes:
    for d in numDisparities:
        plt.figure(figsize=(15,30))
        stereo = cv2.StereoBM_create(numDisparities=d, blockSize=block)
        
        disparityEst = stereo.compute(imgL_gray, imgR_gray)
        errImg = np.zeros(imgL_gray.shape, dtype=np.uint8)
        
#       pixel-wise error calculation
        for r in range(disparityEst.shape[0]):
            for c in range(disparityEst.shape[1]):
                errImg[r][c] = np.sum((disparityEst[r][c] - dispL_gray[r][c]) ** 2)
                
#       total MSE calculation
        mse = np.sum((disparityEst.astype("float") - dispL_gray.astype("float")) ** 2)
        mse /= float(disparityEst.shape[0] * disparityEst.shape[1])
        
        plt.subplot(9,2,i), plt.imshow(disparityEst, 'gray')
        plt.title("Block: " + str(block) + "; Disparity: " + str(d)), plt.xticks([]), plt.yticks([])
        i=i+1
        plt.subplot(9,2,i), plt.imshow(errImg, 'gray')
        plt.title("MSE: " +str(mse)), plt.xticks([]), plt.yticks([])
        i=i+1
        
        
```

Effects of Changing the Window Size:

   * In general as the window size (block size) increases, we get smoother disparity maps with less detail and less noise.

1. Edges: We lose edges as we increase the block size. Block size 5 identifies the edges better than block size 9 and 10 but there is too much noise.
    
2. Depth: Error on the disparity map increases more rapidly for smaller depths as we increase the block size. 
    e.g. The difference in number of errors we get from the shallowest part by increasing the window size from 5 to 9 is larger than the difference in number of errors we get from the deepest part.
    
3. Texture: Increasing the block size affects the high texture areas more than the low texture areas. As we increase the block size texture becomes less obvious and the image becomes smoother. 
    
Effects of Changing the Disparity:
    
   * In this algorithm changing the disparity number increases the disparity search range. Since the cameras that took the two images seem to be pretty close and the objects in the image are not too close to the the cameras, a smaller disparity range works better in general.

1. Edges: Changing the disparity doesn't seem to have a large effect on the edges. 
    
2. Depth: Shallow areas are affected more by the changes in disparity. The depth calculated in shallow areas decrease more than the depth calculated in deeper areas when going from a small disparity to a large disparity. e.g. Shallow areas on disparity level 16 show significantly better quality of depth than disparity level 48, but the difference in quality isn't as obvious in deeper areas.
    
3. Texture: Increasing the disparitu makes the texture more evident if the area has a lot of texture. It doesn't affect the low texture areas significantly.

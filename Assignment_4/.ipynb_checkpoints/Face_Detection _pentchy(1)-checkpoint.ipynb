```python
def createDataMatrix(images):
    ''' 
    Allocate space for all images in one data matrix. 
        The size of the data matrix is
        ( w  * h  * 3, numImages )
        
        where,
         
        w = width of an image in the dataset.
        h = height of an image in the dataset.
        3 is for the 3 color channels.
        '''
   
    numImages = len(images)
    sz = images[0].shape
    data = np.zeros((numImages, sz[0] * sz[1] * sz[2]), dtype=np.float32)
    # for grayscale
    #data = np.zeros((numImages, sz[0] * sz[1]), dtype=np.float32)
    for i in range(0, numImages):
        image = images[i].flatten()
        data[i,:] = image
     
    return data

def createNewFace(*args):
    # Start with the mean image
    output = average_face

    # Add the eigen faces with the weights
    for i in range(0, NUM_EIGEN_FACES):
        '''
        OpenCV does not allow slider values to be negative. 
        So we use weight = sliderValue - MAX_SLIDER_VALUE / 2
        ''' 
        sliderValues[i] = cv2.getTrackbarPos("Weight" + str(i), "Trackbars");
        weight = sliderValues[i] - MAX_SLIDER_VALUE/2
        output = np.add(output, eigen_faces[i] * weight)

    # Display Result at 2x size
    output = cv2.resize(output, (0,0), fx=2, fy=2)
    plt.figure(figsize = (10,10))
    plt.subplot(111)
    plt.imshow(output, cmap='gray')
    plt.title("Result"), plt.xticks([]), plt.yticks([])
```

PCA to represent the Eigenvector representation of our training data

```python
def eigenvectors_and_PCA(dataset):
    image_shape = dataset[0].shape
    
    # Create data matrix with our whole dataset for PCA
    data_matrix = createDataMatrix(dataset)
    
    # Compute the eigenvectors from the stack of images created
    # Max PCA components = 10 -> 10 samples per dataset
    #mean, eigenvectors = cv2.PCACompute(data_matrix, mean=None, maxComponents=10)

    pca = PCA()
    X_transformed = pca.fit_transform(data_matrix)
    
    eigen_vectors = pca.components_
    eigen_values = pca.explained_variance_
    
    #eig_vals, eig_vecs = np.linalg.eig(cov_mat)

    print ("Eigenvectors \n%s" %eigen_vectors)
    print ("")
    print ("Eigenvectors Array Size:" + str(eigen_vectors.shape) + "\n")
    print ("\nEigenvalues \n%s" %eigen_values)
    
    # Make a list of (eigenvalue, eigenvector) tuples
    eigen_pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range(len(eigen_values))]

    # Sort the (eigenvalue, eigenvector) tuples from high to low
    eigen_pairs.sort()
    eigen_pairs.reverse()
    
    # Use the Eigen Values to plot the Explained Variance per Component in PCA
    # (In theory and in practice we have 10 components in our PCA as our data set is 10)
        
    tot = sum(eigen_values)
    var_exp = [(i / tot)*100 for i in sorted(eigen_values, reverse=True)]
    cum_var_exp = np.cumsum(var_exp)

    nb_components = range(1, len(eigen_pairs) + 1)
    plt.plot(nb_components, var_exp, '-o', color="green", label="Explained Var.")
    plt.plot(nb_components, cum_var_exp, '-o', color="blue", label="Cumulative Exp. Var.")
    plt.xlabel("Component #")
    plt.ylabel("Explained Variance")
    plt.legend()
    plt.show()    
```

### Discussion: Do you need all the vectors to represent the data?
No you don't. In fact, the explained variance, taken from the eigenvalues, gives us an idea of how much information each component of PCA gives us. As we see above when we compute the explained variance, most of the information is held by the first component, some in the second. After the fourth component, barely any information is left thus we can only keep the first few components. There is quite a difference when computing grayscale vs. color. In color, the information is more dispersed; in grayscale, most of the information is concentrated in the first three components.


### Original Pixel Space (Nearest Neighbor) with L2 Distance

```python
import sys

# L2 Distance (Eucledian) between two vectors
def l2_eucledian_distance(X, Y):
    return np.linalg.norm(X-Y)

def get_global_dataset(directories):
    dataset = []
    for directory in directories:
        for file_path in os.listdir("./dataset/" + directory):
            if (not file_path.startswith(".")):
                dataset.append([cv2.resize(cv2.imread("./dataset/" + directory + "/" + file_path), (128, 128)), directory])
    return dataset

directories = ["face1", "face2", "face3"]
# New dataset of all samples + labels
global_dataset = np.array(get_global_dataset(directories))
print ("Verify that the global dataset is a 30x2 matrix: " + str(global_dataset.shape) + "\n")

def distance_comparator(global_dataset):
    for item in global_dataset:
        smallest_distance = sys.float_info.max
        smallest_dist_label = "none"
        for compared_item in global_dataset:
            # don't compare the image to itself...
            if (not item == compared_item):
                temp_distance = l2_eucledian_distance(item[0], compared_item[0])
                if (temp_distance < smallest_distance):
                    smallest_distance = temp_distance
                    smallest_dist_label = compared_item[1]
        print ("Image of label: \"" + str(item[1]) + "\" is closest to an image of label: \"" + smallest_dist_label+ "\"")
        if (not item[1] != smallest_dist_label):
            ("This image is outside of its cluster!")

distance_comparator(global_dataset)
```

### Eigenspace (Nearest Neighbor) with L2 Distance

```python
# Eigenspace
def get_eigenvectors(dataset, label):
    # Create data matrix for each dataset for PCA
    data_matrix = createDataMatrix(dataset)
    
    # Compute the eigenvectors from the stack of images created
    # Max PCA components = 10 -> 10 samples per dataset
    _, eigenvectors = cv2.PCACompute(data_matrix, mean=None, maxComponents=10)
    
    dataset = []
    for e_v in eigenvectors:
        dataset.append([e_v, label])
    
    return np.array(dataset)

eigen_vectors_face_one = get_eigenvectors(face_one[0], face_one[1])
eigen_vectors_face_two = get_eigenvectors(face_two[0], face_two[1])
eigen_vectors_face_three = get_eigenvectors(face_three[0], face_three[1])
# Create a eigenspace global dataset
eigenspace_dataset = np.concatenate((eigen_vectors_face_one, eigen_vectors_face_two, eigen_vectors_face_three))
distance_comparator(eigenspace_dataset)
```
